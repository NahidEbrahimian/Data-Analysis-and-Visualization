```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#banana width = 4
#banana length = 8
#apple width =6
#apple length =6

N = 1000
std = 0.5
bananas_width = np.random.normal(4, std, N)
bananas_length = np.random.normal(8, std, N)

apples_width = np.random.normal(6, std, N)
apples_length = np.random.normal(6, std, N)

width = np.concatenate((bananas_width, apples_width))
length = np.concatenate((bananas_length, apples_length))
# print(width.shape)
X = np.array([width, length]).T
# x_ = np.stack(

bananas_labels = np.zeros(N, dtype = 'int')
apples_labels = np.ones(N, dtype = 'int')

Y = np.concatenate((bananas_labels, apples_labels))
```

```python
plt.scatter(X[0 : N, 0], X[0 : N , 1], label='apples')
plt.scatter(X[N :, 0], X[N: , 1], label='bananas')
plt.legend()
plt.xlabel('Length')
plt.ylabel('Width')
```

## Bar chart with error bars


```python
labels = ['bananas', 'apples', 'bananas', 'apples']

x = [1, 1.5, 2.5, 3.0]


height = [bananas_width.mean(),apples_width.mean(),bananas_length.mean(),apples_length.mean()]
yerr = [bananas_width.std(),apples_width.std(),bananas_length.std(),apples_length.std()]

height = [bananas_width.mean(),apples_width.mean(),bananas_length.mean(),apples_length.mean()]
yerr = [bananas_width.std(),apple_width.std(),bananas_length.std(),apples_length.std()]

color = ['tab:blue', 'tab:blue', 'tab:orange', 'tab:orange']

plt.bar(x = x,
        height=height,
        color=color,
        width=0.4,
        error_kw={'lw': 4},
        yerr=yerr)

plt.xticks(ticks=x, labels=labels)
plt.title('Fruit length VS width')

plt.grid()

```

## Box diagram


```python

plt.boxplot(bananas_length,
            patch_artist=True, boxprops={'facecolor': 'tab:blue'},
            positions=[0], labels=['bananas'], widths=[0.15])

plt.boxplot(apples_length,
            patch_artist=True, boxprops={'facecolor': 'tab:orange'},
            positions=[0.3], labels=['apples'], widths=[0.15])

plt.ylabel('length')
plt.xlim(-0.2, 0.5)
plt.title('Length Distribution ')
```

## Feature generation and selection

### 01- Aspect ratio

```python
# fea = 'L/W'
# fruits[fea] = lenght / width

fruits1 = bananas_length / bananas_width
fruits2 = apples_length / apples_width

plt.hist(fruits1,
         bins=30, density=True, color='tab:blue', label='bananas',)

plt.hist(fruits2,
         bins=30, density=True, color='tab:orange', label='apples',)

plt.legend()
plt.title('L/W Distribution ')
plt.xlabel('L/W ratio')
```

### 02- the measure of area


```python
# fea = 'area'
# fruits[fea] = fruits['length']*fruits['width']

fruits1 = bananas_length * bananas_width
fruits2 = apples_length * apples_width

plt.hist(fruits1, histtype='step', lw=4,
         bins=30, density=True, color='tab:blue', label='bananas')

plt.hist(fruits2, histtype='step', lw=4,
         bins=30, density=True, color='tab:orange', label='apples')

plt.title('Area Distribution ')
plt.legend()
plt.xlabel('Area')
```

### 03- Perimeter

```python
# fea = 'circumference'
# fruits[fea] = 2*(fruits['length']+fruits['width'])

fruits1 = (apples_width + apples_length) * 2
fruits2 = (bananas_width + bananas_length) * 2

plt.hist(fruits1, histtype='step', lw=4,
         bins=30, density=True, color='tab:blue', label='bananas')

plt.hist(fruits2, histtype='step', lw=4,
         bins=30, density=True, color='tab:orange', label='apples')

plt.title('circumference Distribution ')
plt.legend()
plt.xlabel('Circumference')
```
