```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

Num_sampels = 500

Balloon_Width = np.random.normal(35, 10, Num_sampels)
Balloon_Length = np.random.normal(35, 10, Num_sampels)
Balloon_Weight = np.random.normal(10, 0.1, Num_sampels)

Melon_Width = np.random.normal(15, 5, Num_sampels)
Melon_Length = np.random.normal(25, 5, Num_sampels)
Melon_Weight = np.random.normal(2500, 1000, Num_sampels)

Width_Feature = np.concatenate((Balloon_Width, Melon_Width))
Length_Feature = np.concatenate((Balloon_Length, Melon_Length))
Weight_Feature = np.concatenate((Balloon_Weight, Melon_Weight))

X = np.array([Width_Feature, Length_Feature, Weight_Feature]).T

# X[:, 0] = (X[:, 0] - X[:, 0].min()) / (X[:, 0].max() - X[:, 0].min())
# X[:, 1] = (X[:, 1] - X[:, 1].min()) / (X[:, 1].max() - X[:, 1].min())
# X[:, 2] = (X[:, 2] - X[:, 2].min()) / (X[:, 2].max() - X[:, 2].min())

Balloon_labels = np.zeros(Num_sampels, dtype = 'int')
Melon_labels = np.ones(Num_sampels, dtype = 'int')

Y = np.concatenate((Balloon_labels, Melon_labels))
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(X[0 : Num_sampels, 0], X[0 : Num_sampels , 1], X[0 : Num_sampels , 2], c = 'red', marker = '*')
ax.scatter(X[Num_sampels :, 0], X[Num_sampels: , 1], X[Num_sampels :, 2], c = 'orange', marker = 'o')

ax.set_xlabel('Widths')
ax.set_ylabel('Length')
ax.set_zlabel('Weight')

plt.show()
```
